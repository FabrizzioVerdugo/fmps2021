Primero creamos la funcion "señal" que realiza las tareas pedidas y retorna un vector con la señal x y el muestreo fs.

```python
def señal(nombre):
    a = wv.read(nombre)
    fs = a[0]
    x = a[1]
    t=np.arange(0,len(x)/fs,1/fs)
    fig, ax = plt.subplots(1,1,figsize=(18,4))
    o1=ax.plot(t,x,'g-',label='$x$ para $N=$ y $f_s=')
    ax.legend()
    ax.set_xlabel('t[s]')
    return(x,fs)
```
